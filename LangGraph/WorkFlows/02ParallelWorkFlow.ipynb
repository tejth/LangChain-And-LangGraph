```python
from langgraph.graph import StateGraph , START , END
from typing import TypedDict
from IPython.display import Image
```

```python
class BatsManState(TypedDict):
    runs:int
    fours:int
    sixes:int
    balls:int 
    
    strike_rate: float
    bp:float
    bpb:float
    
    
```

```python
def calculate_SR(state:BatsManState)->BatsManState:
    result = round( (state['runs']/state['balls'])*100,2)
    
    return {'strike_rate':result}

def boundary_per_balls(state:BatsManState)->BatsManState:
    total_boundaries = state['fours']+state['sixes']
    result= round(state['balls']/total_boundaries,2)
    
    return {'bpb':result}

def boundary_percent(state:BatsManState)->BatsManState:
    result = round(((4*state['fours'] + 6*state['sixes'])/state["runs"])*100,2)
    
    return {"bp":result}
    
    
    
```

```python
#Define a graph
graph = StateGraph(BatsManState)

#Create Nodes
graph.add_node("calculate_SR", calculate_SR)
graph.add_node("boundary_percent", boundary_percent)
graph.add_node("boundary_per_balls",boundary_per_balls)

#Adding edges
graph.add_edge(START,"calculate_SR")
graph.add_edge(START,"boundary_per_balls")
graph.add_edge(START,"boundary_percent")

graph.add_edge("calculate_SR",  END)
graph.add_edge("boundary_per_balls",END)
graph.add_edge("boundary_percent",END)

#Compiler Graph
workflow = graph.compile()

#Execute Graph
result_state  = workflow.invoke({
    "runs":180,
    "fours":5,
    "sixes":2,
    "balls":101
})

print(result_state)
```
