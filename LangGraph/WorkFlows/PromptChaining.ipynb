```python
from langgraph.graph import StateGraph , START , END
from typing import TypedDict
```

```python
class AppState(TypedDict):
    topic:str
    summary:str
    post:str
```

```python
def create_summary(state:AppState)->AppState:
    return {"summary": f"Summary of topic: {state["topic"]}"}

def generate_post(state:AppState)->AppState:
    return {"post":f"LinkedIn post using {state["summary"]}"}
```

```python
#Define graph
graph = StateGraph(AppState)

#Create nodes
graph.add_node("create_summary",create_summary)
graph.add_node("generate_post",generate_post)


#Add Edges
graph.add_edge(START,"create_summary")
graph.add_edge("create_summary", "generate_post")
graph.add_edge("generate_post",END)


#Compile the Graph
workflow = graph.compile()


#Execute the graph
result = workflow.invoke({
    "topic":"Role of Ai!"
})

print(result)
```
